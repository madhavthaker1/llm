```python
from crewai import Agent, Task, Crew


planning_agent = Agent(
    role="Senion Python Developer",
    goal="""Create a detailed plan to solve this task. This plan will be used by a junior developer to write the code. 
    You are also responsible for providing detailed feedback back to the junior developer on the quality of their code.""",
    backstory="You are an experienced programmer with strong Python skills.",
    allow_code_execution=False,
    allow_delegation=True,
    verbose=True,
    llm=llm
)

# Create an agent with code execution enabled
coding_agent = Agent(
    role="Junior Python Developer",
    goal="""Take the plan given to you by the senion python developer and implement it in python using OOP priciples. 
    You will also take any feedback and incorporate it into your code.""",
    backstory="You are an experienced programmer with strong Python skills.",
    allow_code_execution=True,
    allow_delegation=False,
    verbose=True,
    llm=llm
)

# Create a task that requires code execution
coding_task = Task(
    description="Code a sudoku solver.",
    agent=coding_agent,
    expected_output='Python Script'
)

# Create a task that requires code execution
review_task = Task(
    description="Review written code and provide feedback.",
    agent=planning_agent,
    expected_output='Outlined feedback'
)

# Create a crew and add the task
analysis_crew = Crew(
    agents=[planning_agent, coding_agent],
    tasks=[coding_task, review_task],
    verbose=True
)

# Execute the crew
result = analysis_crew.kickoff()

print(result)
```

```python
from crewai import Crew, Process, Agent, Task

# Define your agents
researcher = Agent(
  role='Researcher',
  goal='Conduct foundational research',
  backstory='An experienced researcher with a passion for uncovering insights'
)
analyst = Agent(
  role='Data Analyst',
  goal='Analyze research findings',
  backstory='A meticulous analyst with a knack for uncovering patterns'
)
writer = Agent(
  role='Writer',
  goal='Draft the final report',
  backstory='A skilled writer with a talent for crafting compelling narratives'
)

research_task = Task(description='Gather relevant data...', agent=researcher, expected_output='Raw Data')
analysis_task = Task(description='Analyze the data...', agent=analyst, expected_output='Data Insights')
writing_task = Task(description='Compose the report...', agent=writer, expected_output='Final Report')

# Form the crew with a sequential process
report_crew = Crew(
  agents=[researcher, analyst, writer],
  tasks=[research_task, analysis_task, writing_task],
  process=Process.sequential
)

# Execute the crew
result = report_crew.kickoff()
```
